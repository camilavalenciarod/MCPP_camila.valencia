[10 puntos] Escriba un programa para resolver el siguiente problema

Usted tiene cuatro archivos de texto independientes, con la información de 63 estudiantes. En todos los archivos están los nombres y apellidos (en adelante, "nombres") de los 63 estudiantes y en todos los archivos la información se encuentra ordenada alfabéticamente por apellido o nombre (irrelevante), pero cada archivo tiene información diferente: el archivo 1 ("archivo1.txt") tiene nombres y cédula; el archivo 2 ("archivo2.txt") tiene nombres y e-mail; el archivo 3 ("archivo3.txt") tiene nombres y calificación de evaluación #1; y el archivo 4 ("archivo4.txt") tiene nombres y calificación de evaluación #2. 

En todos los archivos los nombres están separados de la demás información por una coma. Su tarea consiste en crear, a partir de esos cuatro archivos, una variable llamada "estudiantes". Esta será asignada a un diccionario cuyas llaves sean las cédulas de los 63 estudiantes, y sus respectivos valores sean a su vez diccionarios. Cada cédula tendrá como valor un diccionario con cuatro llaves, de manera que al final la variable "estudiantes" tendrá la siguiente forma: 

{<cédula1>: {"Nombres": <nombres>, "e-mail": <e-mail>, "evaluación #1": <evaluación #1>, "evaluación #2": <evaluación 2>}, <cédula2>: {"Nombres": <nombres>, "e-mail": <e-mail>, "evaluación #1": <evaluación #1>, "evaluación #2": <evaluación 2>}, ... , <cédula63>: {"Nombres": <nombres>, "e-mail": <e-mail>, "evaluación #1": <evaluación #1>, "evaluación #2": <evaluación 2>}}. 

(Tenga en cuenta que los signos <> se utilizan para referirse a información que irá en ese lugar, pero no de manera literal. Las comillas, desde luego, sí hacen referencia a contenido literal.)

```python
#Abrir los 4 archivos y poderlos usar mas de una vez

#trabajar el archivo 1

#hacer una lista de cuplas que contengan el nombre y la cedula abriendo el archivo 1
archivo1=[] #crear una lista vacia
with open("archivo1.txt") as f:
    for i in f: #por cada linea del archivo de texto
        nombre, cedula = i.split(",") #separar los elementos por la coma
        archivo1.append ((nombre, cedula)) #agrega esos elemmentos separados como cuplas a la lista vacia 

#Hacer una lista que solo contenga las cedulas

def toma_cedulas (archivo1): #define una funcion
    cedulas=[ ]
    for cupla in archivo1: #por cada elemento de la lista archivo 1
         cedulas.append ( cupla[1]) #agrega a la lista vacia el segundo elemento de cada cupla del archivo1
    return cedulas #genera una lista con solo las cedulas de los estudiantes
cedulas=toma_cedulas(archivo1)
cedulas=[l.rstrip() for l in cedulas]
print(cedulas)

#Hacer una lista que contenga solo los nombres
def toma_nombre (archivo1):
    nombres=[]
    for i in archivo1:
         nombres.append ( i[0])
    return nombres
nombres=toma_nombre(archivo1)

print(nombres)

#hacer una lista de cuplas que contengan el nombre y el correo abriendo el archivo 2
archivo2=[] 
with open('archivo2.txt') as f:
    for i in f:
        nombre, email = i.split(",")
        archivo2.append ((nombre, email))


#Hacer una lista que solo contenga las emails
def toma_emails (archivo2):
    emails=[]
    for cupla in archivo2:
         emails.append ( cupla[1])
    return emails
emails= toma_emails(archivo2)
emails=[l.rstrip() for l in emails]
print(emails)
   
#hacer una lista de cuplas que contengan el nombre y la calificacion 1 abriendo el archivo 3
archivo3=[] 
with open('archivo3.txt') as f:
    for i in f:
        nombre, calificacion1 = i.split(",")
        archivo3.append ((nombre, calificacion1))

#Hacer una lista que solo contenga las calificaciones 1
def toma_calificacion1 (archivo3):
    calificaciones1=[]
    for cupla in archivo3:
         calificaciones1.append (cupla[1])
    return calificaciones1
calificaciones1= toma_calificacion1(archivo3)
calificaciones1=[l.rstrip() for l in calificaciones1]
print(calificaciones1)

#hacer una lista de cuplas que contengan el nombre y la calificacion 2 abriendo el archivo 4
archivo4=[] 
with open("archivo4.txt") as f:
    for i in f:
        nombre, calificacion2 = i.split(",")
        archivo4.append ((nombre, calificacion2))

#Hacer una lista que solo contenga las calificaciones 2
def toma_calificacion2 (archivo4):
    calificaciones2=[ ]
    for cupla in archivo4:
         calificaciones2.append ( cupla[1])
    return calificaciones2
calificaciones2= toma_calificacion2(archivo4)
calificaciones2=[l.rstrip() for l in calificaciones2]
print(calificaciones2)


#Ya tengo las 5 diferentes listas que son: nombres, cedulas, correos, calificaciones1 y calificaciones 2. 

#Ahora hare el diccionario que contiene nombres, correos, y las calificaiones

estudiantes={ } #hacer un diccionario vacio
def crear_estudiantes ( cedulas, nombres, emails, calificaciones1, calificaciones2):
    for i in list(range(len(cedulas))):
        estudiantes [ cedulas[i]] = [{"Nombres": nombres[i], "e-mail": emails[i],"evaluación #1": calificaciones1[i], "evaluación #2": calificaciones2[i]}]  #va agregando elemetos al diccionario vacio de la forma cedula:{}, como una base de datos en excel donde la primera llave es la cedula y los valores de esta llave son un diccionario en el cual esta compuesto de llaves (nombre, correo, calificaion1, calificacion2) con cada uno de los datos de cada uno de  los estudiantes.
    return estudiantes
estudiantes = crear_estudiantes (cedulas, nombres, emails, calificaciones1, calificaciones2)
print(estudiantes)
```
