Vamos a hacer "scraping" a esta página: http://listas.20minutos.es/lista/500-mejores-peliculas-de-la-historia-segun-la-revista-empire-297804/, que contiene un listado de las 500 mejores películas de la historia según la revista Empire.

Su tarea consiste en crear un "Pandas dataframe" que contenga 500 filas (una por película) y las siguientes columnas:
- Ránking (#)
- Título
- Año
- Director
- Reparto
- Sinopsis

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import time
import pandas as pd
import re
```

```python
base_peliculas= "http://listas.20minutos.es/lista/500-mejores-peliculas-de-la-historia-segun-la-revista-empire-297804/"
index_html = urlopen(base_peliculas)
index = BeautifulSoup(index_html, "lxml")
```

```python
peliculas_datos=index.find_all("h3")
```

```python
peliculas_datos
```

```python
secciones = index.find_all("div", {"class": "descripcion_elemento_larga"})
```

```python
secciones
```

```python
ranking=[]
for x in range(0,len(peliculas_datos)):
        fila=peliculas_datos[x].get_text()   
        numero=re.search('\d+',fila).group()
        ranking.append(numero)
```

```python
ranking
```

```python
titulo=[]
for x in range(0,len(peliculas_datos)):
        fila=peliculas_datos[x].get_text()   
        titulos=re.search('.+\w+',fila).group()
        titulo.append(titulos)
```

```python
titulo
```

```python
ano=[]

for x in range(0,len(secciones)):
        fila=secciones[x].get_text()   
        anos = re.search('\d+',fila).group()
        ano.append(anos)
```

```python
ano
```

```python
director=[]

for x in range(0,len(secciones)):
        fila=secciones[x].get_text()   
        directores=re.findall('Director: (\w+ \w+)',fila)
        director.append(directores)
```

```python
director
```

```python
reparto=[]

for x in range(0,len(secciones)):
        fila=secciones[x].get_text()   
        repartos=re.findall('Reparto:(\w+ \w+)',fila)
        reparto.append(repartos)
```

```python
reparto
```

```python
sinopsis=[]

for x in range(0,len(secciones)):
        fila=secciones[x].get_text()   
        sinop=re.findall('Sinopsis:[\D+]+', fila)
        sinopsis.append(sinop)
```

```python
sinopsis
```

```python
l1=pd.DataFrame({'Ranking': ranking})
l2=pd.DataFrame({'Título':titulo})
l3=pd.DataFrame({'Año':ano})
l4=pd.DataFrame({'Director':director})
l5=pd.DataFrame({'Reparto':reparto})
l6=pd.DataFrame({'Sinopsis':sinopsis})
datos=pd.concat([l1,l2,l3,l4,l5,l6], ignore_index=False, axis=1)
datos
```
